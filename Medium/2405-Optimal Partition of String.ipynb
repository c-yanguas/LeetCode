```python
class Solution:
    def partitionString(self, s: str) -> int:
        num_partitions = 1
        temp_sub_str = ""
        for char in s:
            if char not in temp_sub_str:
                temp_sub_str = temp_sub_str + char
            else:
                temp_sub_str = char
                num_partitions = num_partitions + 1
        return num_partitions




solution = Solution()
solution.partitionString("abacaba"), solution.partitionString("ssssss")
```
